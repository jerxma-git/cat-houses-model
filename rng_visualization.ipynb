```python
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon

from customrng import CustomRandomGenerator

```

```python
# init
rng = CustomRandomGenerator(12345)
```

```python
# uniform
uniform_samples = [rng.uniform() for _ in range(100000)]

plt.figure(figsize=(10, 6))
plt.hist(uniform_samples, bins=50, density=True, alpha=0.6, color='b')
plt.title('Uniform Distribution [0,1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.axhline(1, color='r', linestyle='--', label='Theoretical')
plt.legend()
plt.show()
```

```python
mu, sigma = 0.5, 0.2

#  
normal_samples = [rng.normal(mu, sigma) for _ in range(100000)]

#  
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(normal_samples, bins=50, density=True, alpha=0.6, color='g')
plt.plot(x, pdf, 'r-', lw=2, label='Theoretical PDF')
plt.title(f'Normal Distribution (={mu}, ={sigma})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
scale = 2
exp_samples = [rng.exponential(scale) for _ in range(100000)]

# Theoretical PDF
x = np.linspace(0, 10, 100)
pdf = expon.pdf(x, scale=scale)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(exp_samples, bins=50, density=True, alpha=0.6, color='pink', range=(0, 10))
plt.plot(x, pdf, 'r-', lw=2, label='Theoretical PDF')
plt.title(f'Exponential Distribution (={1/scale})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
mu, sigma, a, b = 8, 2, 1, 10
trunc_samples = [rng.truncated_normal(mu, sigma, a, b) for _ in range(100000)]

# Theoretical PDF
x = np.linspace(a, b, 100)
pdf = norm.pdf(x, mu, sigma)
pdf = pdf / (norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma))

# Plot
plt.figure(figsize=(10, 6))
plt.hist(trunc_samples, bins=50, density=True, alpha=0.6, color='purple')
plt.plot(x, pdf, 'r-', lw=2, label='Theoretical PDF')
plt.title(f'Truncated Normal (={mu}, ={sigma}, [{a},{b}])')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
